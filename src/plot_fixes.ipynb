```python
import sys, os

MARIPOSA_PATH = "/home/yizhou7/mariposa/"

sys.path.append(MARIPOSA_PATH + "src")
os.chdir(MARIPOSA_PATH)

from utils.plot_utils import *
import matplotlib.pyplot as plt
import numpy as np

categories = ["Mariposa TO", "Mariposa QU", "Verus TO", "Verus QU"]
total_counts = np.array([447, 98, 39, 31])

no_proof = np.array([3, 1, 2, 1])
fixable_1 = np.array([276, 52, 25, 22])
fixable_2 = np.array([35, 14, 4, 3])

# Stacked bar plot
fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.6

# Add total count bar
ax.bar(
    categories,
    total_counts - fixable_1 - fixable_2,
    bottom=fixable_1 + fixable_2,
    label="Unfixed",
    color="white",
    edgecolor="black",
    linewidth=1,
)

ax.bar(
    categories,
    fixable_2,
    bottom=fixable_1,
    label="Doubleton Fixable",
    color="#803E75",
    alpha=0.7,
    edgecolor="black",
    linewidth=1,
)

# ax.bar(categories, no_proof, label="1-no-proof", color="red")
ax.bar(
    categories,
    fixable_1,
    label="Singleton Fixable",
    color="#803E75",
    alpha=1,
    edgecolor="black",
    linewidth=1,
)


# Labels and legend
ax.set_ylabel("Unstable Query Count")
# ax.set_title("Breakdown of Proof Failures by Category with Total Counts")
ax.legend()

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("fig/debugger/fixable.pdf", dpi=300)
```
